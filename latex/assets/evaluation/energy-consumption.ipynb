# Energy Consumption

```python
import matplotlib.pyplot as plt
```

```python
k3s_file = './measurement/k3s.log'
aws_file = './measurement/aws.log'
ioFog_file = './measurement/ioFog.log'

number_of_points = 3600 # we have an hour of points (3600)
```

```python
def read_data(file):
    data = []
    with open(file) as f:
        for i, line in enumerate(f):
            data.append((i, float(line)))
            if i == number_of_points:
                break
    return data

k3s_data = read_data(k3s_file)
aws_data = read_data(aws_file)
ioFog_data = read_data(ioFog_file)
```

```python
plt.plot(*zip(*k3s_data), "-", label='k3s')
plt.plot(*zip(*aws_data), "-", label='AWS Greengrass')
plt.plot(*zip(*ioFog_data), "-", label='ioFog')
plt.ylabel("Watts")
plt.xlabel("Time")
plt.xticks([])
plt.legend(loc='lower left')
plt.savefig(fname=f'./plots/energy-consumption.png')
```
